```python
import os

dir_path = "output/scenes/17"

files = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
files
```

```python
import mmcv
import numpy as np
import cv2
import skimage


for j, file in enumerate(files):
    if j <= 78:
        continue
        
    video_reader = mmcv.VideoReader('output/scenes/17/' + file)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    video_writer = cv2.VideoWriter('output/fixed/17/' + file, fourcc, video_reader.fps,(video_reader.width, video_reader.height))
    
    for i, frame in enumerate(video_reader):
        if i == 0:
            a = frame
            continue
        else:
            b = frame
            color_diff = skimage.color.deltaE_ciede2000(a, b)
            np.putmask(a[:, :, 0], color_diff > 5, 0)
            np.putmask(a[:, :, 1], color_diff > 5, 0)
            np.putmask(a[:, :, 2], color_diff > 5, 0)
        video_writer.write(a)

    if video_writer:
        video_writer.release()  
```
